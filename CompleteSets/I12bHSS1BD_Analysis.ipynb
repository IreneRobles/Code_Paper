```julia



function BF_HSS1_WT()
    exps = all_ehn
expvector = [exps[exps[!,:Rep].==ii, :] for ii in unique(exps[!,:Rep])]
BFs_gene = calculatesumary(expvector...,limit = 1)

ylimit = 0.34
u = 0.01


TEST = BF_test(BFs_gene, "$intronname", :N_TSS, [
        ["Rad21KO_DMSO", "WT_DMSO"],
        ["Rad21KO_DMSO-LPS", "WT_DMSO-LPS"],
        ["Rad21KO_BD1-LPS", "WT_BD1-LPS"],
        ["Rad21KO_BD2-LPS", "WT_BD2-LPS"],
        ["WT_DMSO", "WT_DMSO-LPS"],
        ["WT_BD1-LPS", "WT_DMSO-LPS"],
        ["WT_BD2-LPS", "WT_DMSO-LPS"],
        ["WT_BD2-LPS", "WT_BD1-LPS"],
        ["Rad21KO_DMSO", "Rad21KO_DMSO-LPS"],
        ["Rad21KO_BD1-LPS", "Rad21KO_DMSO-LPS"],
        ["Rad21KO_BD2-LPS", "Rad21KO_DMSO-LPS"],
        ["Rad21KO_BD1-LPS", "Rad21KO_BD2-LPS"],
        ])

CSV.write("$AnalysisName/"*"$intronname"*"_BFs.csv", BFs_gene)
CSV.write("$AnalysisName/"*"$intronname"*"_BFs_TEST.csv", TEST)
   title("WT")

BFs_gene_sub = BFs_gene[BFs_gene[!, :Genotype].== "WT", :]

Seaborn.stripplot(data = Pandas.DataFrame(BFs_gene_sub), y = "BF", x = "Timepoint", hue = "Genotype", dodge = true, palette = ["darkgray" ],order = ["DMSO", "DMSO-LPS","BD1-LPS","BD2-LPS"])
Seaborn.boxplot(data = Pandas.DataFrame(BFs_gene_sub), y = "BF", x = "Timepoint", hue = "Genotype", dodge = true, palette = ["darkgray"],order = ["DMSO", "DMSO-LPS","BD1-LPS","BD2-LPS"])
ylabel("Enh burst freq.")
xlabel("Treatment")
plt.tight_layout()

h1,h2,h3, h4 = 0.2, 0.24, 0.28, 0.25

ii = 0
h = h1
xy = [-0+ii, 1+ii]; xy2 = [Statistics.mean(xy) , h+0.005]
plt.plot(xy, [h, h], c = "black")
pval = TEST[5, :padj]
plt.annotate(NoLongerProblems.transform_pvalue_in_stars(pval), xy = xy2, ha= "center")
h = h2
xy = [1, 2]; xy2 = [Statistics.mean(xy) , h+0.005]
plt.plot(xy, [h, h], c = "black")
pval = TEST[6, :padj]
plt.annotate(NoLongerProblems.transform_pvalue_in_stars(pval), xy = xy2, ha= "center")

ii=+1
h = h3
xy = [1, 3]; xy2 = [Statistics.mean(xy) , h+0.005]
plt.plot(xy, [h, h], c = "black")
pval = TEST[7, :padj]
plt.annotate(NoLongerProblems.transform_pvalue_in_stars(pval), xy = xy2, ha= "center")

ii=+1
h = h4
xy = [2, 3]; xy2 = [Statistics.mean(xy) , h+0.005]
plt.plot(xy, [h, h], c = "black")
pval = TEST[8, :padj]
plt.annotate(NoLongerProblems.transform_pvalue_in_stars(pval), xy = xy2, ha= "center")

ylim(0, ylimit);legend_removal();pretty_axes2();line075black();squareplot();xticks(rotation=90)

end
function BF_HSS1_Rad21()
    exps = all_ehn
exp1 = exps[exps[!,:Rep].==1, :]
exp2 = exps[exps[!,:Rep].==2, :];
exp4 = exps[exps[!,:Rep].==4, :];

ylimit = 0.34
u = 0.01


BFs_gene = calculatesumary(exp1, exp2, exp4,limit = hss1)
TEST = BF_test(BFs_gene, "$intronname", :N_TSS, [
        ["Rad21KO_DMSO", "WT_DMSO"],
        ["Rad21KO_DMSO-LPS", "WT_DMSO-LPS"],
        ["Rad21KO_BD1-LPS", "WT_BD1-LPS"],
        ["Rad21KO_BD2-LPS", "WT_BD2-LPS"],
        ["WT_DMSO", "WT_DMSO-LPS"],
        ["WT_BD1-LPS", "WT_DMSO-LPS"],
        ["WT_BD2-LPS", "WT_DMSO-LPS"],
        ["WT_BD2-LPS", "WT_BD1-LPS"],
        ["Rad21KO_DMSO", "Rad21KO_DMSO-LPS"],
        ["Rad21KO_BD1-LPS", "Rad21KO_DMSO-LPS"],
        ["Rad21KO_BD2-LPS", "Rad21KO_DMSO-LPS"],
        ["Rad21KO_BD1-LPS", "Rad21KO_BD2-LPS"],
        ])

CSV.write("$AnalysisName/"*"$intronname"*"_BFs.csv", BFs_gene)
CSV.write("$AnalysisName/"*"$intronname"*"_BFs_TEST.csv", TEST)

title("Rad21KO")

BFs_gene_sub = BFs_gene[BFs_gene[!, :Genotype].== "Rad21KO", :]

Seaborn.stripplot(data = Pandas.DataFrame(BFs_gene_sub), y = "BF", x = "Timepoint", hue = "Genotype", dodge = true, palette = ["red"],
order = ["DMSO", "DMSO-LPS","BD1-LPS","BD2-LPS"])
Seaborn.boxplot(data = Pandas.DataFrame(BFs_gene_sub), y = "BF", x = "Timepoint", hue = "Genotype", dodge = true, palette = ["red"],
order = ["DMSO", "DMSO-LPS","BD1-LPS","BD2-LPS"])
ylabel("Enh burst freq.")
xlabel("Treatment")
plt.tight_layout()

h1,h2,h3, h4 = 0.3, 0.28, 0.32, 0.15

ii = 0
h = h1
xy = [-0+ii, 1+ii]; xy2 = [Statistics.mean(xy) , h+0.005]
plt.plot(xy, [h, h], c = "black")
pval = TEST[9, :padj]
plt.annotate(NoLongerProblems.transform_pvalue_in_stars(pval), xy = xy2, ha= "center")



h = h2
xy = [1, 2]; xy2 = [Statistics.mean(xy) , h+0.005]
plt.plot(xy, [h, h], c = "black")
pval = TEST[10, :padj]
plt.annotate(NoLongerProblems.transform_pvalue_in_stars(pval), xy = xy2, ha= "center")

ii=+1
h = h3
xy = [1, 3]; xy2 = [Statistics.mean(xy) , h+0.005]
plt.plot(xy, [h, h], c = "black")
pval = TEST[11, :padj]
plt.annotate(NoLongerProblems.transform_pvalue_in_stars(pval), xy = xy2, ha= "center")

ii=+1
h = h4
xy = [2, 3]; xy2 = [Statistics.mean(xy) , h+0.005]
plt.plot(xy, [h, h], c = "black")
pval = TEST[12, :padj]
plt.annotate(NoLongerProblems.transform_pvalue_in_stars(pval), xy = xy2, ha= "center")



ylim(0, ylimit);legend_removal();pretty_axes2();line075black();squareplot();xticks(rotation=90)

plt.tight_layout()

end


function plot_burstsizes(df; limit = 0, r = "r2", maxntoconsider = 50, GENE = "", max = 150,  palette = ["red", "darkgray"], order = ["DMSO", "DMSO-LPS","BD1-LPS","BD2-LPS"])
    set = df
    if GENE == "Hprt" ||GENE == "Fh1"
        set = set[set[!, :Timepoint].!="0i", :]
        set = set[set[!, :Timepoint].!="8i", :]
        set[!, :Timepoint] = [parse(Int, ii)*60 for ii in set[!, :Timepoint]]
    end
    wt_data = set
    wt_data = tss_data(wt_data; limit = limit)
    wt_data = wt_data[wt_data[!,"TSS1_"*r].>limit, :]
    reps = split_by(wt_data, :Rep)
    
    wt_means = vcat([calculate_medians(reps[ii], "TSS1_"*r; col = "Timepoint", rep = ii) for ii in keys(reps)]...)
    
     reps = split_by(wt_data, :Rep)
    
    
    for ii in  keys(reps)
        samp = split_by(reps[ii], :Sample) 
        for aa in   keys(samp)
            if nrow(samp[aa]) > maxntoconsider
             samp[aa] =  samp[aa][shuffle(1:nrow(samp[aa])), :][1:maxntoconsider, :]
            end
        end
        reps[ii] =  vcat([samp[aa] for aa in   keys(samp)]...)
    end
    
    reps = vcat([reps[aa] for aa in   keys(reps)]...)
   
    
    Seaborn.stripplot(data = Pandas.DataFrame(reps), x = "Timepoint", y = "TSS1_"*r, hue = "Genotype", dodge = true, palette = palette, order = order)
    Seaborn.boxplot(data = Pandas.DataFrame(reps), x = "Timepoint", y = "TSS1_"*r, hue = "Genotype", showfliers = false, palette = palette,dodge = true, order = order)
   
    pretty_axes2();line075black();squareplot();xticks(rotation=90)

    
    title(GENE)
    ylabel("Burst Size ("*r*")")
    
    xlabel("Treatment")
    legend_removal()
    ylim(0, max)
    
    return reps


end

function add_tests_bs(genedata, compsplot, hi; u = 3)
    
t = R"""
tb <- $genedata
tb$Timepoint <- as.factor(tb$Sample)
tb$Rep <- as.factor(tb$Rep); 
tb$TSS1_r2 <- as.numeric(tb$TSS1_r2); 
aov.s = aov(TSS1_r2 ~ Timepoint +Rep ,data=tb)  #do the analysis of variance
test <- TukeyHSD(aov.s)
a = test$Timepoint
    write.csv(a, "BS_test.csv")
"""
t = CSV.read("BS_test.csv", DataFrames.DataFrame)
    
    pvals =  vcat([t[t[!,:Column1].==ii, :] for ii in compsplot]...)[!,"p adj"]
    

cords = [
        [0.75, 1.25],
        [-0.25, 0.75],
        [0.25, 1.25]
    ]
for ii in 1:length(pvals)
    h = hi[ii]
    plot(cords[ii], [h, h], c = "Black")
    annotate(NoLongerProblems.transform_pvalue_in_stars(pvals[ii]), xy =[mean(cords[ii]), h+u], ha = "center", va = "center")
    
end
end

function BS_WTRad21_Il12b()

limity = 170
u = 10
genesize_gene = plot_burstsizes(all_gene, GENE ="$intronname", limit = il12b, palette = [ "darkgray","red"], order = ["DMSO", "DMSO-LPS","BD1-LPS","BD2-LPS"]
)
filename = "$AnalysisName/"*"$intronname"*"_BS_Test.csv"
R"""
tb <- $genesize_gene
tb$Timepoint <- as.factor(tb$Sample)
tb$Rep <- as.factor(tb$Rep); 
tb$TSS1_r2 <- as.numeric(tb$TSS1_r2); 
aov.s = aov(TSS1_r2 ~ Timepoint +Rep ,data=tb)  #do the analysis of variance
test <- TukeyHSD(aov.s)
a = test$Timepoint
write.csv(a, $filename)
"""
t = CSV.read(filename, DataFrames.DataFrame)

s1 = "WT_DMSO-LPS"
s2 = "Rad21KO_DMSO-LPS"
p = 1
if sum(t[!,:Column1].== s1*"-"*s2)==1
    p = t[t[!,:Column1].== s1*"-"*s2, "p adj"]
else
    p = t[t[!,:Column1].== s2*"-"*s1, "p adj"]
end

h = 150

cords = [0.75,1.25]
plot(cords, [h, h], c = "Black")
annotate(NoLongerProblems.transform_pvalue_in_stars(p[1]), xy =[mean(cords), h+u], ha = "center", va = "center")

s1 = "WT_BD1-LPS"
s2 = "Rad21KO_BD1-LPS"
p = 1
if sum(t[!,:Column1].== s1*"-"*s2)==1
    p = t[t[!,:Column1].== s1*"-"*s2, "p adj"]
else
    p = t[t[!,:Column1].== s2*"-"*s1, "p adj"]
end

h = 140

cords = [1.75,2.25]
plot(cords, [h, h], c = "Black")
annotate(NoLongerProblems.transform_pvalue_in_stars(p[1]), xy =[mean(cords), h+u], ha = "center", va = "center")


s1 = "WT_BD2-LPS"
s2 = "Rad21KO_BD2-LPS"
p = 1
if sum(t[!,:Column1].== s1*"-"*s2)==1
    p = t[t[!,:Column1].== s1*"-"*s2, "p adj"]
else
    p = t[t[!,:Column1].== s2*"-"*s1, "p adj"]
end

h = 130
cords = [2.75,3.25]
plot(cords, [h, h], c = "Black")
annotate(NoLongerProblems.transform_pvalue_in_stars(p[1]), xy =[mean(cords), h+u], ha = "center", va = "center")

ylim(0, limity);line075black(); title("WT and Rad21KO")

end

function BS_WT_Il12b()

limity = 170
u = 10

filename = "$AnalysisName/"*"$intronname"*"_BS_Test.csv"

t = CSV.read(filename, DataFrames.DataFrame)


allsam = all_gene[all_gene[!,:Genotype].=="WT", :]
genesize_gene = plot_burstsizes(allsam, GENE ="$intronname", limit = il12b, palette = ["darkgray"], order = ["DMSO", "DMSO-LPS","BD1-LPS","BD2-LPS"])

s1 = "WT_DMSO-LPS"
s2 = "WT_BD1-LPS"
p = 1
if sum(t[!,:Column1].== s1*"-"*s2)==1
    p = t[t[!,:Column1].== s1*"-"*s2, "p adj"]
else
    p = t[t[!,:Column1].== s2*"-"*s1, "p adj"]
end

h = 130
u = 10
cords = [1,2]
plot(cords, [h, h], c = "Black")
annotate(NoLongerProblems.transform_pvalue_in_stars(p[1]), xy =[mean(cords), h+u], ha = "center", va = "center")


s1 = "WT_DMSO-LPS"
s2 = "WT_BD2-LPS"
p = 1
if sum(t[!,:Column1].== s1*"-"*s2)==1
    p = t[t[!,:Column1].== s1*"-"*s2, "p adj"]
else
    p = t[t[!,:Column1].== s2*"-"*s1, "p adj"]
end

h = 150
cords = [1,3]
plot(cords, [h, h], c = "Black")
annotate(NoLongerProblems.transform_pvalue_in_stars(p[1]), xy =[mean(cords), h+u], ha = "center", va = "center")

ylim(0, limity);line075black(); title("WT")


s1 = "WT_BD1-LPS"
s2 = "WT_BD2-LPS"
p = 1
if sum(t[!,:Column1].== s1*"-"*s2)==1
    p = t[t[!,:Column1].== s1*"-"*s2, "p adj"]
else
    p = t[t[!,:Column1].== s2*"-"*s1, "p adj"]
end

h = 120
cords = [2,3]
plot(cords, [h, h], c = "Black")
annotate(NoLongerProblems.transform_pvalue_in_stars(p[1]), xy =[mean(cords), h+u], ha = "center", va = "center")

ylim(0, limity);line075black();squareplot()

end

function BS_Rad21_Il12b()

limity = 170
u = 10

filename = "$AnalysisName/"*"$intronname"*"_BS_Test.csv"

t = CSV.read(filename, DataFrames.DataFrame)

allsam = all_gene[all_gene[!,:Genotype].=="Rad21KO", :]
genesize_gene = plot_burstsizes(allsam, GENE ="$intronname", limit = il12b, palette = ["red"], order = ["DMSO", "DMSO-LPS","BD1-LPS","BD2-LPS"])

s1 = "Rad21KO_DMSO-LPS"
s2 = "Rad21KO_BD1-LPS"
p = 1
if sum(t[!,:Column1].== s1*"-"*s2)==1
    p = t[t[!,:Column1].== s1*"-"*s2, "p adj"]
else
    p = t[t[!,:Column1].== s2*"-"*s1, "p adj"]
end

h = 130
u = 10
cords = [1,2]
plot(cords, [h, h], c = "Black")
annotate(NoLongerProblems.transform_pvalue_in_stars(p[1]), xy =[mean(cords), h+u], ha = "center", va = "center")


s1 = "Rad21KO_DMSO-LPS"
s2 = "Rad21KO_BD2-LPS"
p = 1
if sum(t[!,:Column1].== s1*"-"*s2)==1
    p = t[t[!,:Column1].== s1*"-"*s2, "p adj"]
else
    p = t[t[!,:Column1].== s2*"-"*s1, "p adj"]
end

h = 150
cords = [1,3]
plot(cords, [h, h], c = "Black")
annotate(NoLongerProblems.transform_pvalue_in_stars(p[1]), xy =[mean(cords), h+u], ha = "center", va = "center")

ylim(0, limity);line075black(); title("WT")


s1 = "Rad21KO_BD1-LPS"
s2 = "Rad21KO_BD2-LPS"
p = 1
if sum(t[!,:Column1].== s1*"-"*s2)==1
    p = t[t[!,:Column1].== s1*"-"*s2, "p adj"]
else
    p = t[t[!,:Column1].== s2*"-"*s1, "p adj"]
end

h = 120
cords = [2,3]
plot(cords, [h, h], c = "Black")
annotate(NoLongerProblems.transform_pvalue_in_stars(p[1]), xy =[mean(cords), h+u], ha = "center", va = "center")

ylim(0, limity);line075black();squareplot();title("Rad21KO")


    plt.tight_layout()
    
end


function BS_WTRad21_HSS1()

limity = 100
u = 6

genesize_gene = plot_burstsizes(all_ehn, GENE ="$ehnname", limit = hss1, palette = ["darkgray","red" ], order = ["DMSO-LPS","BD1-LPS","BD2-LPS"]
)
filename = "$AnalysisName/"*"$ehnname"*"_BS_Test.csv"
R"""
tb <- $genesize_gene
tb$Timepoint <- as.factor(tb$Sample)
tb$Rep <- as.factor(tb$Rep); 
tb$TSS1_r2 <- as.numeric(tb$TSS1_r2); 
aov.s = aov(TSS1_r2 ~ Timepoint +Rep ,data=tb)  #do the analysis of variance
test <- TukeyHSD(aov.s)
a = test$Timepoint
write.csv(a, $filename)
"""
t = CSV.read(filename, DataFrames.DataFrame)

s1 = "WT_DMSO-LPS"
s2 = "Rad21KO_DMSO-LPS"
p = 1
if sum(t[!,:Column1].== s1*"-"*s2)==1
    p = t[t[!,:Column1].== s1*"-"*s2, "p adj"]
else
    p = t[t[!,:Column1].== s2*"-"*s1, "p adj"]
end


h = 80

cords = [-0.25,0.25]
plot(cords, [h, h], c = "Black")
annotate(NoLongerProblems.transform_pvalue_in_stars(p[1]), xy =[mean(cords), h+u], ha = "center", va = "center")

s1 = "WT_BD1-LPS"
s2 = "Rad21KO_BD1-LPS"
p = 1
if sum(t[!,:Column1].== s1*"-"*s2)==1
    p = t[t[!,:Column1].== s1*"-"*s2, "p adj"]
else
    p = t[t[!,:Column1].== s2*"-"*s1, "p adj"]
end


h = 80

cords = [0.75,1.25]
plot(cords, [h, h], c = "Black")
annotate(NoLongerProblems.transform_pvalue_in_stars(p[1]), xy =[mean(cords), h+u], ha = "center", va = "center")

s1 = "WT_BD2-LPS"
s2 = "Rad21KO_BD2-LPS"
p = 1
if sum(t[!,:Column1].== s1*"-"*s2)==1
    p = t[t[!,:Column1].== s1*"-"*s2, "p adj"]
else
    p = t[t[!,:Column1].== s2*"-"*s1, "p adj"]
end

h = 80

cords = [1.75,2.25]
plot(cords, [h, h], c = "Black")
annotate(NoLongerProblems.transform_pvalue_in_stars(p[1]), xy =[mean(cords), h+u], ha = "center", va = "center")

ylim(0, limity);line075black(); title("WT and Rad21KO"); squareplot()

end

function BS_WT_HSS1()
  filename = "$AnalysisName/"*"$ehnname"*"_BS_Test.csv"  
limity = 100
u = 6
t = CSV.read(filename, DataFrames.DataFrame)


allsam = all_ehn[all_ehn[!,:Genotype].=="WT", :]
genesize_gene = plot_burstsizes(allsam, GENE ="$intronname", limit = hss1, palette = ["darkgray"], order = ["DMSO", "DMSO-LPS","BD1-LPS","BD2-LPS"])

s1 = "WT_DMSO-LPS"
s2 = "WT_BD1-LPS"
p = 1
if sum(t[!,:Column1].== s1*"-"*s2)==1
    p = t[t[!,:Column1].== s1*"-"*s2, "p adj"]
else
    p = t[t[!,:Column1].== s2*"-"*s1, "p adj"]
end

h = 80

cords = [1,2]
plot(cords, [h, h], c = "Black")
annotate(NoLongerProblems.transform_pvalue_in_stars(p[1]), xy =[mean(cords), h+u], ha = "center", va = "center")


s1 = "WT_DMSO-LPS"
s2 = "WT_BD2-LPS"
p = 1
if sum(t[!,:Column1].== s1*"-"*s2)==1
    p = t[t[!,:Column1].== s1*"-"*s2, "p adj"]
else
    p = t[t[!,:Column1].== s2*"-"*s1, "p adj"]
end

h = 90
cords = [1,3]
plot(cords, [h, h], c = "Black")
annotate(NoLongerProblems.transform_pvalue_in_stars(p[1]), xy =[mean(cords), h+u], ha = "center", va = "center")

ylim(0, limity);line075black(); title("WT")


s1 = "WT_BD1-LPS"
s2 = "WT_BD2-LPS"
p = 1
if sum(t[!,:Column1].== s1*"-"*s2)==1
    p = t[t[!,:Column1].== s1*"-"*s2, "p adj"]
else
    p = t[t[!,:Column1].== s2*"-"*s1, "p adj"]
end

h = 70
cords = [2,3]
plot(cords, [h, h], c = "Black")
annotate(NoLongerProblems.transform_pvalue_in_stars(p[1]), xy =[mean(cords), h+u], ha = "center", va = "center")

ylim(0, limity);line075black();squareplot()
end

function BS_Rad21_HSS1()
    filename = "$AnalysisName/"*"$ehnname"*"_BS_Test.csv"
limity = 100
u = 6
t = CSV.read(filename, DataFrames.DataFrame)

allsam = all_ehn[all_ehn[!,:Genotype].=="Rad21KO", :]
genesize_gene = plot_burstsizes(allsam, GENE ="$ehnname", limit = hss1, palette = ["red"], order = ["DMSO", "DMSO-LPS","BD1-LPS","BD2-LPS"])

s1 = "Rad21KO_DMSO-LPS"
s2 = "Rad21KO_BD1-LPS"
p = 1
if sum(t[!,:Column1].== s1*"-"*s2)==1
    p = t[t[!,:Column1].== s1*"-"*s2, "p adj"]
else
    p = t[t[!,:Column1].== s2*"-"*s1, "p adj"]
end

h = 80

cords = [1,2]
plot(cords, [h, h], c = "Black")
annotate(NoLongerProblems.transform_pvalue_in_stars(p[1]), xy =[mean(cords), h+u], ha = "center", va = "center")


s1 = "Rad21KO_DMSO-LPS"
s2 = "Rad21KO_BD2-LPS"
p = 1
if sum(t[!,:Column1].== s1*"-"*s2)==1
    p = t[t[!,:Column1].== s1*"-"*s2, "p adj"]
else
    p = t[t[!,:Column1].== s2*"-"*s1, "p adj"]
end

h = 90
cords = [1,3]
plot(cords, [h, h], c = "Black")
annotate(NoLongerProblems.transform_pvalue_in_stars(p[1]), xy =[mean(cords), h+u], ha = "center", va = "center")


s1 = "Rad21KO_BD1-LPS"
s2 = "Rad21KO_BD2-LPS"
p = 1
if sum(t[!,:Column1].== s1*"-"*s2)==1
    p = t[t[!,:Column1].== s1*"-"*s2, "p adj"]
else
    p = t[t[!,:Column1].== s2*"-"*s1, "p adj"]
end

h = 70
cords = [2,3]
plot(cords, [h, h], c = "Black")
annotate(NoLongerProblems.transform_pvalue_in_stars(p[1]), xy =[mean(cords), h+u], ha = "center", va = "center")

ylim(0, limity);line075black();squareplot();title("Rad21KO")

filename = "$AnalysisName/"*"$ehnname"*"_BS_Test.svg"


plt.tight_layout()
    
end
```
