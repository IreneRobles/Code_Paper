```julia
using RCall
R"""
###################################
# DESeq2 GROseq enhancer analysis #
###################################

# Back in December 2015 Gopu did an analysis on the 
# Sergi GROseq macrophage reads. However he only compared
# WT to Rad21KO macrophages. In order to make a figure stating that changes in gene expression upon LPS occur through changes in transcription, I needed comparisons accross timepoints.
# As I had Gopu original code, I had just to track down the files in the ax4, which were in:

#.     - /csc/analysis/Merkenshlager/SergiCuartero/Groseq_Dec2015/Merged

# And then I adapted the code to make the comparisons I needed

library(dplyr)
library(DESeq2);
library(IRanges)
library(GenomicRanges)


load("../Databases/Cuartero2018/GROseq/OstuneEnhModified_Counts_Dec2015.RData")
deseq <- function(sample1, sample2){
    GROEnhCounts <- EnhancerSummaryCounts
colnames(GROEnhCounts) <- gsub(".merged.dupMarkd.bam", "", colnames(GROEnhCounts))
    group <- factor(substr(colnames(GROEnhCounts), 1, 3))  # for all samples
    group <- relevel(group, sample2)

    condition <- data.frame(condition = group, row.names = colnames(GROEnhCounts))
    dds <- DESeqDataSetFromMatrix(countData = GROEnhCounts, colData = condition, 
    design = ~condition)

    GROWTdds <- DESeq(dds)
    subdds_results <- results(GROWTdds, contrast=c("condition",sample1,sample2),alpha=0.1)
    subdds_results<- cbind(as.data.frame(OstuniEnhGR_modified), as.data.frame(subdds_results))
    write.table(subdds_results, paste0("../Databases/Cuartero2018/GROseq/GROseq_",sample1, "__vs__", sample2, "__DESeq2_OtsuniEnhancers.csv"),sep=",",quote=F,row.names=F)
    return(subdds_results)
}

deseq("WT1","WT0");
deseq("WT6","WT0");
deseq("FL0","WT0");
deseq("FL1","WT1");
deseq("FL6","WT6");
deseq("WT6","WT1");
deseq("FL1","FL0");
deseq("FL6","FL0");
deseq("FL6","FL1");
""";
```
