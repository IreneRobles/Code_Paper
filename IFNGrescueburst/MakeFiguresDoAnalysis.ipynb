# IFNG pre-reatment and Il12b and Egr2 transcriptional bursts

## Load data and rin BS test

```julia
include("HiddenCode.jl")
include("../Figure1d_Il12b_BSBF_WT/HiddenCode_BSBF.jl")

gene1_name = "Il12b"
gene1 = CSV.read("../../Code_Paper/CompleteSets/GeneData/Il12b_intron_IFNG.csv", DataFrames.DataFrame, missingstring = "NA")
enh1_name = "HSS1"
enh1 = CSV.read("../../Code_Paper/CompleteSets/GeneData/HSS1_IFNG.csv", DataFrames.DataFrame, missingstring = "NA")
gene2_name = "Egr2"
gene2 = CSV.read("../../Code_Paper/CompleteSets/GeneData/Egr2_intron_IFNG.csv", DataFrames.DataFrame, missingstring = "NA")
enh2_name = "Egr2 Enh"
enh2 = CSV.read("../../Code_Paper/CompleteSets/GeneData/Egr2_enh_IFNG.csv", DataFrames.DataFrame, missingstring = "NA")

g1 = make_sumaries(gene1, gene1_name, int = "r2", limit = 1)
g2 = make_sumaries(gene2, gene2_name, int = "r2", limit =1)
e11 = make_sumaries(enh1[enh1[!,:Rep].==1, :], enh1_name, int = "r2", limit = 1)
e12 = make_sumaries(enh1[enh1[!,:Rep].==2, :], enh1_name, int = "r2", limit = 1);e12[!,:Rep] = [2 for ii in 1:nrow(e12)]
e1 = vcat(e11,e12)
e2 = make_sumaries(enh2, enh2_name, int = "r2")

do_bs_tests(gene1, gene1_name)
do_bs_tests(gene2, gene2_name)
do_bs_tests(enh1, enh1_name, limits = [1,1])
do_bs_tests(enh2, enh2_name);
```

## Burst Frequency

```julia
subplot(2,2,1)
title("Il12b")
BF_fig_90(g1; pal = ["darkgray","red"], hs = [0.03, 0.05, 0.1, 0.05])
subplot(2,2,2)
title("HSS1")
BF_fig_90(e1; pal = ["darkgray","red"], hs = [0.25, 0.2, 0.22, 0.25])


subplot(2,2,3)
title("Egr2")
BF_fig_90(g2; pal = ["darkgray","red"], hs = [0.05, 0.15, 0.45, 0.30], tim = 60)
subplot(2,2,4)
title("Egr2_Enh")

BF_fig_90(e2; pal = ["darkgray","red"], hs = [0.1, 0.13, 0.15, 0.1], tim = 60)

plt.tight_layout()
```
