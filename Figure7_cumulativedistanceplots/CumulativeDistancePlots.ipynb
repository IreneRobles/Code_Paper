# Cumulative distance plots between genes and enhancers

P value = Kolgomorov-Smirnov test

## Rad21KO vs WT

```julia
include("../Figure3f_distances/HiddenCode.jl")
include("Figures.jl")
figure(figsize = (15, 7))
suff = "200um_noz"
limit = 10
subplot(2,4,1)
show_distances2("../CompleteSets/linkedlocus/","Egr2_intron","Egr2_enh",suff, limit = limit)
subplot(2,4,2)
show_distances2("../CompleteSets/linkedlocus/","Prdm1_intron","Prdm1_enh",suff,  limit = limit)
subplot(2,4,3)
show_distances2("../CompleteSets/linkedlocus/","Il12b_intron","HSS1",suff, limit = limit, limit_enh =0)
subplot(2,4,4)
show_distances2("../CompleteSets/linkedlocus/","Peli1_intron","Enh",suff, limit = limit)
limit = 1
subplot(2,4,5)
show_distances2("../CompleteSets/linkedlocus/","Egr2_intron","Egr2_enh",suff, limit = limit)
subplot(2,4,6)
show_distances2("../CompleteSets/linkedlocus/","Prdm1_intron","Prdm1_enh",suff,  limit = limit)
subplot(2,4,7)
show_distances2("../CompleteSets/linkedlocus/","Il12b_intron","HSS1",suff, limit = limit, limit_enh =0)
subplot(2,4,8)
show_distances2("../CompleteSets/linkedlocus/","Peli1_intron","Enh",suff, limit = limit)

legend()
plt.tight_layout()
savefigwithtext("GeneEnhdistancesCumplot___link"*suff*".svg")
```
