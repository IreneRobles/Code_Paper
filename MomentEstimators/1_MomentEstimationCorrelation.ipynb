# Moment Estimation of transcriptional parameters

From Bagnall et al. 2020

\begin{align}
   \sigma^2 &= variance \\\
   \mu &= mean \\\
   b_m &= {moment burst size}\\\
   f_m &= {moment burst frequency}\\\
   Burst fraction &=  \frac{k_{on}}{k_{on}+k_{off}} \\\
   Burst Half Life &=  \frac{ln(2)}{k_{off}} \\\
   Burst Size &=  Burst Half Life * k_{tc} = \frac{ln(2)}{k_{off}}* k_{tc} \\\
   b_m &= \frac{\sigma^2}{\mu} \\\
   \\\
   f_m &= \frac{\mu}{b_m-1} \\\
\end{align}
